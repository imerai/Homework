```python
import pandas as pd
```

##### School csv file 

```python
path1 = '/Users/VINESH/Desktop/GitLab/nu-class-homework/HW4/schools_complete.csv'
schools = pd.read_csv(path1, encoding='iso-8859-1', low_memory=False)
schools.head()
```

##### Student csv file 

```python
path2 = '/Users/VINESH/Desktop/GitLab/nu-class-homework/HW4/students_complete.csv'
students = pd.read_csv(path2, encoding='iso-8859-1', low_memory=False)
students.head()
```

##### School district summary

```python
# Get the total number of school filtered by school id assuming if there was a duplicate school id
total_schools = len(schools['School ID'].unique())
# Get the total number of students filtered by student id assuming if there was a duplicate student id
total_students = len(students['Student ID'].unique())
# Total up all school budget
total_budget = schools['budget'].sum()

# Derive average math score
avg_math_score = round((students['math_score'].mean()),2)
# Derive average reading score
avg_read_score = round((students['reading_score'].mean()),2)
# 65 as the passing grade, locate in student file and create a data frame
math_pass_df = students.loc[students['math_score'] >= 65]
# Round it to 2 decimal as percentage
math_pass = round((len(math_pass_df) / total_students * 100),2)
# 65 as the passing grade, locate in student file and create a data frame
read_pass_df = students.loc[students['reading_score'] >= 65]
# Round it to 2 decimal as percentage
read_pass = round((len(read_pass_df) / total_students * 100),2)
# Merge the two data frame to get the overall pass
ovall_pass_df = pd.merge(math_pass_df, read_pass_df, on='Student ID')
# Round it to 2 decimal as percentage
ovall_pass = round((len(ovall_pass_df) / total_students * 100),2)
# Create a dataframe setting all the values from above into a dict
dist_df = pd.DataFrame({'Total Schools': [total_schools], 'Total Students': [total_students], 'Total Budget':
                         [total_budget], 'Average Math Score': [avg_math_score], 'Average Reading Score':
                         [avg_read_score], '% Passing Math': [math_pass], '% Passing Reading': [read_pass],
                         'Overall Passing Rate': [ovall_pass]})
# format budget numberic value in the dict
dist_df['Total Budget'] = dist_df['Total Budget'].map("$ {:,.2f}".format)
dist_df
# This step is not needed
# distr_df_result = distr_df[['Total Schools', 'Total Students', 'Total Budget', 'Average Math Score', 'Average Reading Score',
#                          '% Passing Math', '% Passing Reading', '% Overall Passing Rate']]
# distr_df_result
```

#### School Summary

```python
# Assign the 5 columns of school file to col identifiers
school_tot = schools[['school_name', 'type', 'size', 'budget']]
# Assign a new col to the new dict and assign per student cost
school_tot = school_tot.assign(Per_Student_Budget = (school_tot['budget'] / school_tot['size']))
# Set the columns names
school_tot.columns = ['School name', 'School Type', 'Total Students', 'Total School Budget', 'Per Student Budget']
# set up a dict for average reading and math scores
students_scr = students[['school_name', 'reading_score', 'math_score']]
# Group the data by school name
student_sum = students_scr.groupby(['school_name'])
# Derive the math and reading mean for each grouped school
student_sum_df = pd.DataFrame(student_sum.mean())
# Remove any NA value from the dataframe as a result of the action above and reset index to 0 dropping the origin indexing
student_sum_df.reset_index(inplace=True)
# Set student data columns names
student_sum_df.columns = ['School name', 'Average Reading Score', 'Average Math Score']

# Derive math score as a new dict
students_scr_math = students[['school_name', 'math_score']]
# Get the passing math score
all_math_df = students_scr_math.loc[students_scr_math['math_score'] >= 65]
# Group score by school
math_sum = all_math_df.groupby(['school_name'])
# Accumlate total math by school
math_sum_df = pd.DataFrame(math_sum.count())
# Remove any NA value from the dataframe as a result of the action above and reset index to 0 dropping the origin indexing
math_sum_df.reset_index(inplace=True)
# Assign column names
math_sum_df.columns = ['School name', 'Passed Math Count']

# Derive reading score as a new dict
students_scr_read = students[['school_name', 'reading_score']]
# Get the passing reading score
all_read_df = students_scr_read.loc[students_scr_read['reading_score'] >= 65]
# Group score by school
read_sum = all_read_df.groupby(['school_name'])
# Accumlate total math by school
read_sum_df = pd.DataFrame(read_sum.count())
# Remove any NA value from the dataframe as a result of the action above and reset index to 0 dropping the origin indexing
read_sum_df.reset_index(inplace=True)
# Assign column names
read_sum_df.columns = ['School name', 'Passed Read Count']

# Merge the two data frame data on the school name as new dict
new_sch1 = pd.merge(school_tot, student_sum_df, on='School name')
# Take the new dict and add the math data to a new dict
new_sch2 = pd.merge(new_sch1, math_sum_df, on='School name')
# Take the new dict and add the reading data to a new dict
new_sch3 = pd.merge(new_sch2, read_sum_df, on='School name')
# Now calculate percentage for counts
new_sch3['% Passing Math'] = round((new_sch3['Passed Math Count'] / new_sch3['Total Students'] * 100),2)
new_sch3['% Passing Reading'] = round((new_sch3['Passed Read Count'] / new_sch3['Total Students'] * 100),2)
new_sch3['% Overall Passing Rate'] = (new_sch3['% Passing Math'] + new_sch3['% Passing Reading'])/2
# # Format the dollar fields and set index to school name
# new_sch3 = new_sch3.set_index('School name')
# new_sch3['Total School Budget'] = new_sch3['Total School Budget'].map("$ {:,.2f}".format)
# new_sch3['Per Student Budget'] = new_sch3['Per Student Budget'].map("$ {:,.2f}".format)
# # Display the table
# new_sch3

# Assign meaningful titles to columns and use that for indexing 
# This step is necessary for steps below
sch_sum_df = new_sch3[['School name', 'School Type', 'Total Students', 'Total School Budget', 'Per Student Budget',
                      'Average Reading Score', 'Average Math Score','% Passing Reading', '% Passing Math',
                      '% Overall Passing Rate']]
dspl_sch_sum = sch_sum_df.set_index('School name')
dspl_sch_sum['Total School Budget'] = dspl_sch_sum['Total School Budget'].map("$ {:,.2f}".format)
dspl_sch_sum['Per Student Budget'] = dspl_sch_sum['Per Student Budget'].map("$ {:,.2f}".format)
dspl_sch_sum
```

##### Top Performing Schools (By Passing Rate)

```python
top_schools = sch_sum_df.sort_values('% Overall Passing Rate', ascending=False)
#top_schools = new_sch3.sort_values('% Overall Passing Rate', ascending=False)
#print(top_schools)
top = top_schools.set_index('School name')
top_schools['Total School Budget'] = top_schools['Total School Budget'].map("$ {:,.2f}".format)
top_schools['Per Student Budget'] = top_schools['Per Student Budget'].map("$ {:,.2f}".format)
top_five = top_schools.head(5)
top_five
```

##### Bottom Performing Schools (By Passing Rate)

```python
bottom_schools = sch_sum_df.sort_values('% Overall Passing Rate', ascending=True)
bottom = bottom_schools.set_index('School name')
bottom['Total School Budget'] = bottom['Total School Budget'].map("$ {:,.2f}".format)
bottom['Per Student Budget'] = bottom['Per Student Budget'].map("$ {:,.2f}".format)
bottom_five = bottom.head(5)
bottom_five
```

##### Math Scores by Grade

```python
# Get math and reading scores, will be used here and in next step
grade_both = students.groupby(['school_name','grade'])
#grade_both.head()
grade_both_df = pd.DataFrame(grade_both.mean())
#grade_both_df.head()
grade_both_df.drop('Student ID', axis=1, inplace=True)
# grade_both_df.head()
grade_both_df.reset_index(inplace=True)
#grade_both_df.head()
piv_math = grade_both_df.pivot(index='school_name', columns='grade', values='math_score') #pivot function,very convenient
#piv_math.head()
math_grade = piv_math[['9th', '10th', '11th', '12th']]
math_grade
```

##### Reading Scores by Grade

```python
piv_read = grade_both_df.pivot(index='school_name', columns='grade', values='reading_score')
read_grade = piv_read[['9th', '10th', '11th', '12th']]
read_grade
```

##### Scores by School Spending

```python
print(dspl_sch_sum['Per Student Budget'].min())
print(dspl_sch_sum['Per Student Budget'].max())
```

```python
bins = [0, 585, 610, 635, 660]
labels_budg = ['<$ 585', '$ 585-610', '$ 610-635', '$ 635-660']
# It's important to use .copy(), so the initial df won't be changed
budg_df = sch_sum_df.copy()
budg_df['Spending Ranges (Per Student)'] = pd.cut(budg_df["Per Student Budget"],bins,labels=labels_budg)
budg_df.drop('Per Student Budget', axis=1, inplace=True)
budg_df.drop('Total School Budget', axis=1, inplace=True)
budg_df.drop('Total Students', axis=1, inplace=True)
budg_grouped = budg_df.groupby(['Spending Ranges (Per Student)'])
budg_grouped_df = pd.DataFrame(budg_grouped.mean())
budg_grouped_df
```

##### Scores by School Size

```python
bins = [0, 1000, 3000, 5000]
labels_size = ['Small (<1000)', 'Medium (1000-3000)', 'Large (>3000)']
size_df = sch_sum_df.copy()
size_df['School Size'] = pd.cut(size_df['Total Students'],bins,labels=labels_size)
size_df.drop('Per Student Budget', axis=1, inplace=True)
size_df.drop('Total School Budget', axis=1, inplace=True)
size_df.drop('Total Students', axis=1, inplace=True)
size_grouped = size_df.groupby(['School Size'])
size_grouped_df = pd.DataFrame(size_grouped.mean())
size_grouped_df
```

##### Scores by School Type

```python
school_type = sch_sum_df.copy()
type_grouped = school_type.groupby('School Type')
type_grouped_df = pd.DataFrame(type_grouped.mean())
type_grouped_df.drop('Total School Budget', axis=1, inplace=True)
type_grouped_df.drop('Total Students', axis=1, inplace=True)
type_grouped_df['Per Student Budget'] = type_grouped_df['Per Student Budget'].map("$ {:,.2f}".format)
type_grouped_df
```
