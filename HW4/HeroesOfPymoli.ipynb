```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# Raw data file
file_to_load = "/Users/VINESH/Desktop/GitLab/nu-class-homework/HW4/purchase_data.csv"

# Read purchasing file and store into pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

```python
# Total number of unique players (based on SN)
total_players = len(purchase_data['SN'].unique())
# Create DF
total_players_df = pd.DataFrame({'Total number of players': [total_players]})
total_players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Number of unique items bought
number_unique_items = len(purchase_data['Item ID'].unique())
# Average price
avg_price = round(purchase_data['Price'].mean(), 2)
# Total number of purchases
tot_purchase = purchase_data['Item ID'].count()
# Total revenue
tot_revenue = purchase_data['Price'].sum()
# Create DF, set and format values
purchase_data_df = pd.DataFrame({'Number of Unique Items': [number_unique_items], 'Average Purchase Price': [avg_price],
                                 'Total Number of Purchases': [tot_purchase], 'Total Revenue': [tot_revenue]})
purchase_data_df['Average Purchase Price'] = purchase_data_df['Average Purchase Price'].map("$ {:,.2f}".format)
purchase_data_df['Total Revenue'] = purchase_data_df['Total Revenue'].map("$ {:,.2f}".format)
purchase_data_dspy = purchase_data_df[['Number of Unique Items', 'Average Purchase Price', 'Total Number of Purchases',
                                   'Total Revenue']]
# Set index to Number of Unique Items
purchase_data_dspy = purchase_data_dspy.set_index('Number of Unique Items')
purchase_data_dspy
```

## Gender Demographics

* Percentage and Count of Male Players

* Percentage and Count of Female Players

* Percentage and Count of Other / Non-Disclosed

```python
# Create a DF with Gender & SN only
gender_df = purchase_data[['SN', 'Gender']]
# Delete all dups
scrub_gender = gender_df.drop_duplicates(['SN'], keep='last')
# Change index to Gender
scrub_gender_idx = scrub_gender.set_index('Gender')
# Use loc. to count all 3 variations
male_count = scrub_gender_idx.loc['Male', 'SN'].count()
female_count = scrub_gender_idx.loc['Female', 'SN'].count()
other_count = scrub_gender_idx.loc['Other / Non-Disclosed', 'SN'].count()
# Count persentages
male_perc = round((male_count/total_players*100), 2)
female_perc = round((female_count/total_players*100), 2)
other_perc = round((other_count/total_players*100), 2)
# Create DF
gender_dspy = {'Gender': ['Male', 'Female', 'Other / Non-Disclose'],
              'Count of Players': [male_count, female_count, other_count],
              'Percentage of Players': [male_perc, female_perc, other_perc]}
gender_dspy_df = pd.DataFrame(gender_dspy)
gender_dspy = gender_dspy_df.set_index('Gender')
gender_dspy
```


## Purchasing Analysis (Gender)

* Create a summary data frame to hold the results

* Optional: give the displayed data cleaner formatting

* Display the summary data frame

## Age Demographics

* Establish bins for ages

* Categorize the existing players using the age bins. Hint: use pd.cut()

* Calculate the numbers and percentages by age group

* Create a summary data frame to hold the results

* Optional: round the percentage column to two decimal points

* Display Age Demographics Table

```python
# Create bins and labels to them
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
# Add new column with bins
purchase_data['Age Bins'] = pd.cut(purchase_data["Age"],age_bins,labels=group_names)
# Group by new column
age_grped = purchase_data.groupby('Age Bins')
# Pick up values
age_purch_count = (age_grped['Age'].count())
age_avg_purch_price = round((age_grped['Price'].mean()), 2)
age_total_purch_value = (age_grped['Price'].sum())
age_df = purchase_data[['SN', 'Age', 'Age Bins']]
# Delete all duplicates in SN
clean_age = age_df.drop_duplicates(['SN'], keep='last')
clean_age_grped = clean_age.groupby('Age Bins')
age_demo_df = pd.DataFrame({'Total Count': clean_age_grped.size(),
                      'Percentage of Players': round(((clean_age_grped.size()/total_players)*100),2)})
age_demo_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below

* Create a summary data frame to hold the results

* Sort the total purchase value column in descending order

* Optional: give the displayed data cleaner formatting

* Display a preview of the summary data frame

```python
# Extract SN and Price, group by SN
price_sn = purchase_data[['SN', 'Price']]
# Group by name
pr_sn_grped = price_sn.groupby(['SN'])
# Add all price grouped by name
price_df = pd.DataFrame(pr_sn_grped.sum())
# Now sort by price
price_sorted = price_df.sort_values('Price', ascending=False)
# Keep the top five spenders
spenders_price = price_sorted.head(5)
# Reset index to SN
spenders_price.reset_index(inplace=True)

# Create a second dataframe with purchase counts
price_count_df = pd.DataFrame(pr_sn_grped.count())
# Set the index to SN
price_count_df.reset_index(inplace=True)
# Merge the two dataframe into one using indexed SN to get name, price and count together
merged_price = pd.merge(spenders_price, price_count_df, on='SN')
# Now give the columns meaningful names
merged_price.columns = ['SN', 'Total Purchase Value', 'Purchase Count']
# Format the price and count vlaues
merged_price['Average Purchase Price'] = round((merged_price['Total Purchase Value'] / merged_price['Purchase Count']),2)
# Set the index to SN
top5result = merged_price.set_index('SN')
# Format the price and value
top5result['Average Purchase Price'] = top5result['Average Purchase Price'].map("$ {:,.2f}".format)
top5result['Total Purchase Value'] = top5result['Total Purchase Value'].map("$ {:,.2f}".format)
# Rearrange the col
topFive = top5result[['Purchase Count', 'Average Purchase Price', 'Total Purchase Value']]
topFive
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns

* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value

* Create a summary data frame to hold the results

* Sort the purchase count column in descending order

* Optional: give the displayed data cleaner formatting

* Display a preview of the summary data frame

```python
# Extract necessary columns for origin data
items = purchase_data[['Item ID','Item Name', 'Price', 'SN']]
# Assign 1 to all row for SN, need to get top items
items = items.assign(SN= 1)
# Assign column names
items.columns = ['Item ID','Item Name', 'Total Purchase Value', 'Purchase Count']
# Group by Name and item ID
items_grouped = items.groupby(['Item ID', 'Item Name'])
# Grouped items total count
items_df = pd.DataFrame(items_grouped.sum())
# Sort by # of purchaces
items_sorted = items_df.sort_values('Purchase Count', ascending=False)
# Sort to get top 5 selling items
popular_items = items_sorted.head(5)
# Assign a new col and put the derived value in it
popular_items = popular_items.assign(Item_Price=(popular_items['Total Purchase Value'] / popular_items['Purchase Count']))
# Format the fields
popular_items['Total Purchase Value'] = popular_items['Total Purchase Value'].map("$ {:,.2f}".format)
popular_items['Item_Price'] = popular_items['Item_Price'].map("$ {:,.2f}".format)
popular_items_dspl = popular_items[['Purchase Count', 'Item_Price', 'Total Purchase Value']]
popular_items_dspl
```
